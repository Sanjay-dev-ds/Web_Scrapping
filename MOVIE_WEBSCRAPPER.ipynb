```python
import requests
from bs4 import BeautifulSoup
import lxml  #working with xml
import pandas as pd
from xlwt import *
```

```python
url = 'https://www.rottentomatoes.com/browse/movies_at_home/sort:popular?page=1'
```

```python
headers = {

    'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36 QIHU 360SE'
 
    }
```

```python
f= requests.get(url,headers)
```

```python
soup = BeautifulSoup(f.content, 'lxml')
```

```python
def get_movie_review(urls) :
  review = []
  for url in urls:
   f= requests.get(url,headers)
   soup = BeautifulSoup(f.content, 'lxml')
   if(soup.find('span',  {'data-qa': 'critics-consensus'}) != None):
        review.append(soup.find('span',  {'data-qa': 'critics-consensus'}).text.strip())
   else:
        review.append(' ')


  return review


   
```

```python
movies = soup.find_all('a',  {'class': 'js-tile-link'})
movie_Date = []  
name = []      
Score = []   
links = []   
review = []                                                   
for movie in movies:
 movie_Date.append(movie.find('span',  {'class': 'smaller'}).text.strip())
 name.append(movie.find('span',  {'class': 'p--small'}).text.strip())
 Score.append(movie.find('score-pairs').get('audiencescore'))
 links.append('https://www.rottentomatoes.com'+str(movie.get('href')))




movie_Date.remove(movie_Date[0])
name.remove(name[0])
Score.remove(Score[0])
links.remove(links[0])
review = get_movie_review(links)



dict = {
    
    'Movie Name':name,
    'Release Date':movie_Date,
    'Audience Score':Score,
    'Movie Links':links,
    'Review' : review
  
}

df = pd.DataFrame(dict)

df.head(10)

```

```python
from pathlib import Path  
filepath = Path('drive/MyDrive/Colab Notebooks/WEB-Scrapper/ROTTEN.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
df.to_csv(filepath,index=False)
```
